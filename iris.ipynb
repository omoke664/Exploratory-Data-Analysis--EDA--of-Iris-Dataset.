Import Necessary Libraries

```python
import pandas as pd  #import Pandas library for data manipulation
import matplotlib.pyplot as plt #for creating visualizations
import numpy as np #for numerical operations
import seaborn as sns #for statistical data visualization
```

Load the dataset

```python
#loading the dataset from scikit-learn
from sklearn.datasets import load_iris 
iris = load_iris()
```

```python
#Create the Pandas DataFrame from the Iris data 
data = pd.DataFrame(data = iris.data, columns = iris.feature_names)
```

```python
#Add the target variable (species) to the DataFrame
data['target'] = iris.target
```

```python
#Display the first few rows of the DataFrame
data.head()
```

```python
#Print the data types of each column
data.info()
```

```python
#Calculate summary statistics for numerical columns
data.describe()
```

the describe() method in Pandas returns a DataFrame containing summary statistics for numerical columns. These statistics typically include:
Central Tendency
mean: indicates the average value of the data.
median: represents the middle value when data is stored.
Dispersion
Standard deviation: measures how spread out the data is from the mean
Range: Difference between the maximum and minimum values.
Interquartile range (IQR): difference between the third and first quartiles, representing the spread of the middle 50% of the data.
Distribution
Percentile: gives information about the distribution of the data.

```python
#check for missing values in the dataset
data.isnull().sum()
```

```python
#create histogram for each numerical feature
data.hist(figsize=(10, 8))
plt.show()
```

```python
#Create box plots for each numerical feature
sns.boxplot(data = data)
plt.show()
```

```python
#Create pairwise scatter plots with color-coded species
sns.pairplot(data, hue='target')
plt.show()
```
